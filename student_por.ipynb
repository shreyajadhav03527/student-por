```python
import pandas as pd
df=pd.read_csv("student-por.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df['G2'].mean()
```

```python
# finding the avg in first case

avg_score=df['G3'].mean()
avg_score
# finding the above avg score
above_avg_students=df[df['G3']>avg_score]
above_avg_students[['G3']]
# count the no of go above avg
above=(df['G3']>avg_score).sum()
print("above avg",above)

# count tne below avg
below=(df['G3']<avg_score).sum()
print("below avg",below)
```

```python
# find out top 5 student
df.sort_values(by='G3',ascending=False).head(5)[['G1','G2','G3']]
# find out the 5 bottam students
df.sort_values(by='G3',ascending=True).head(5)[['G1','G2','G3']]
```

```python
# find how many students are passed and failed
passed=(df['G3']>=10).sum()
print("passed",passed)
failed=(df['G3']<10).sum()
print("failed",failed)
```

```python
# find the student with 0 absences
df[df['absences']==0]
df[df['absences']==0][['G1','G2','G3']]
```

```python
# find the students who imporoved from G1 to G3
df[df['G3']>df['G1']]
df[df['G3']>df['G1']][['G1','G3']]


```

```python
# find the student who performed consisitently among g1,g2,g3
df[(df['G1']==df['G2'])&(df['G2']==df['G3'])]
df[(df['G1']==df['G2'])&(df['G2']==df['G3'])][['G1','G2','G3']]
```

```python
# matplotlib
# matplotlib tools is helpful for visualization
# histogram
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'],bins=10)
plt.xlabel('Final Grade')
plt.ylabel('Number of Students')
plt.title('Histogram of Final Grades')
plt.show()
```

```python
df.columns
```

```python
# barchat
# avg final marks scored by boys and girls
avg_marks_by_gender=df.groupby('sex')['G3'].mean()
avg_marks_by_gender

plt.figure()
avg_marks_by_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')
plt.title('Average Final Grade by Gender')
plt.show()

```

```python
# scatter plot
plt.figure()
plt.scatter(df['studytime'],df['G3'])
plt.xlabel('study time')
plt.ylabel('G3')
plt.title('scatter plot of G1 vs G3')
plt.show()
```

```python
plt.figure()
plt.scatter(df['studytime'],df['G2'])
plt.xlabel('study time')
plt.ylabel('G2')
plt.title('scatter plot of G1 vs G2')
plt.show()
```

```python
#box plot
plt.figure()
plt.boxplot(df['G3'])
plt.xlabel('G3')
plt.ylabel('G3')
plt.title('Box Plot of G3')
plt.show()
```

```python
# line plot
avg_marks=df[['G1','G2','G3']].mean()
plt.figure()
plt.plot(avg_marks.index,avg_marks.values,marker='o')
plt.xlabel("subject")
plt.ylabel("average marks")
plt.title("average marks by subject")
plt.show()

```

```python
# with bar chat we will find the numbere of students who passed and failed
# who failed and passed the final exam
passed=(df['G3']>=10).sum()
print("passed",passed)
failed=(df['G3']<10).sum()
print("failed",failed)


plt.figure()
plt.bar(['passed','failed'],[passed,failed],color=['green','red'])
plt.xlabel("result")
plt.ylabel("number of students")
plt.title("number of students who passed and failed")
plt.show()
```

```python
# histogram- absesence distribution
plt.figure()
plt.hist(df['absences'],bins=10,color='black',edgecolor='white')
plt.xlabel("absences")
plt.ylabel("number of students")
plt.title("absences distribution")
plt.show()

# color option blue red green orange purple skyblue
plt.figure()
plt.hist(df['absences'],bins=10,color='skyblue',edgecolor='red')
plt.xlabel("absences")
plt.ylabel("number of students")
plt.title("absences distribution")
plt.show()
#
```

```python
# we are seeing about seaborn
# which is built on top of matplotlib and it is easier to use and  easier to plot
import seaborn as sns
sns.histplot(df['G3'],bins=10,color='red',edgecolor='white')
plt.show()
```

```python
# avg by boys and girls using seaborn
sns.barplot(x='sex',y='G3',data=df,color='purple')
plt.show()
```

```python
# scatter plot using seaborn
sns.scatterplot(x='studytime',y='G3',data=df,color='red')
plt.show()
```
